Dado um conjunto de dados com categorias (por exemplo, dados de vendas por região), realize uma amostragem estratificada mantendo a proporção de cada categoria na população original.



```python
# Contando quantos registros tem em cada região
contagem_regioes = df_vendas['Região'].value_counts()
print("Distribuição Original da População:")
print(contagem_regioes)

# Definindo tamanho da amostra total
tamanho_amostra = 50

# Aplicando a amostra estratificada
amostra_estratificada = df_vendas.groupby('Região', group_keys=False).apply(
    lambda x: x.sample(frac=tamanho_amostra / len(df_vendas), random_state=42),
)

print("\nDistribuição da Amostra Estratificada:")
print(amostra_estratificada['Região'].value_counts())
```

### Exercício 4: Inferência Estatística
Extraia 100 amostras aleatórias de tamanho 30 e calcule a média de cada amostra. Compare a distribuição dessas médias com a média populacional.


```python
import numpy as np

# Definindo a população
np.random.seed(42)
populacao = np.random.normal(loc=50, scale=10, size=1000)  # População com média 50 e desvio padrão 10

# Definindo os parâmetros que vao ser usados
n_amostras = 100  # Número de amostras
tamanho_amostra = 30  # Tamanho de cada amostra

# Extraindo 100 amostras aleatórias e calculando a média de cada uma
medias_amostrais = [np.mean(np.random.choice(populacao, size=tamanho_amostra, replace=False)) for _ in range(n_amostras)]

# Calculando a média da população
media_populacional = np.mean(populacao)

# Comparando as médias
print(f"Média Populacional: {media_populacional:.2f}")
print(f"Média da Primeira Amostra: {medias_amostrais[0]:.2f}")
print(f"Média da Última Amostra: {medias_amostrais[-1]:.2f}")
print(f"Média das Médias Amostrais: {np.mean(medias_amostrais):.2f}")
```

### Exercício 5: Cálculo de Medidas Básicas
Para um conjunto de dados (por exemplo, preços de produtos), calcule:
- Média
- Mediana
- Moda
- Amplitude
- Variância
- Desvio padrão
- Coeficiente de variação


```python
# Definindo o número de produtos
num_produtos = 50

# Gerando os preços aleatórios para os produtos
np.random.seed(42)  # Para garantir reprodutibilidade
precos = np.random.uniform(10, 100, num_produtos)  # Preços entre 10 e 100

media = precos.mean()
mediana = np.median(precos)
moda = pd.Series(precos).mode()[0]
amplitude = precos.max() - precos.min()
variancia = precos.var()
desvio_padrao = precos.std()
coeficiente_variacao = (desvio_padrao / media) * 100

print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Amplitude: {amplitude:.2f}")
print(f"Variância: {variancia:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")
print(f"Coeficiente de Variação: {coeficiente_variacao:.2f}%")
```


### Exercício 6: Detecção de Outliers (opcional )
Identifique outliers em um conjunto de dados usando:
- Método do Z-score
- Método do IQR (Intervalo Interquartil)
Crie visualizações que destaquem esses outliers.


```python
# Gerando dados com outliers
np.random.seed(42)
dados = np.random.normal(loc=50, scale=10, size=100)
dados = np.append(dados, [120, 130, 10, 5])  # Adicionando outliers

# 1. Método do Z-score
# Calculando o Z-score para cada ponto de dado
z_scores = (dados - np.mean(dados)) / np.std(dados)

# Definindo o limite para outliers (ex: Z-score > 3 ou Z-score < -3)
limite_zscore = 3
outliers_zscore = dados[np.abs(z_scores) > limite_zscore]

# 2. Método do IQR
# Calculando os quartis Q1 e Q3
Q1 = np.percentile(dados, 25)
Q3 = np.percentile(dados, 75)
IQR = Q3 - Q1

# Definindo os limites inferior e superior para outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers_iqr = dados[(dados < limite_inferior) | (dados > limite_superior)]


# Visualização dos outliers
plt.figure(figsize=(12, 6))

# Box plot para visualizar outliers com IQR
plt.subplot(1, 2, 1)
sns.boxplot(x=dados)
plt.title('Box Plot - Detecção de Outliers com IQR')

# Scatter plot para visualizar outliers com Z-score
plt.subplot(1, 2, 2)
plt.scatter(range(len(dados)), dados, color='blue', label='Dados')
plt.scatter(np.where(np.abs(z_scores) > limite_zscore)[0], outliers_zscore, color='red', label='Outliers (Z-score)')
plt.title('Scatter Plot - Detecção de Outliers com Z-score')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Outliers detectados com Z-score: {outliers_zscore}")
print(f"Outliers detectados com IQR: {outliers_iqr}")
```


### Exercício 7: Comparação de Conjuntos (opcional )
Compare duas amostras diferentes (por exemplo, desempenho de dois grupos em um teste) usando medidas de tendência central e dispersão. Determine se há diferenças significativas entre os grupos.


```python
import scipy.stats as stats

# Gerando dados para dois grupos
np.random.seed(42)  # Para reprodutibilidade
grupo1 = np.random.normal(loc=70, scale=10, size=50)  # Grupo 1 com média 70 e desvio padrão 10
grupo2 = np.random.normal(loc=75, scale=12, size=60)  # Grupo 2 com média 75 e desvio padrão 12

# Medidas de tendência central
media_grupo1 = np.mean(grupo1)
media_grupo2 = np.mean(grupo2)
mediana_grupo1 = np.median(grupo1)
mediana_grupo2 = np.median(grupo2)

# Medidas de dispersão
desvio_padrao_grupo1 = np.std(grupo1)
desvio_padrao_grupo2 = np.std(grupo2)
variancia_grupo1 = np.var(grupo1)
variancia_grupo2 = np.var(grupo2)

# Teste t de Student para comparar as médias
t_statistic, p_value = stats.ttest_ind(grupo1, grupo2)

# Imprimindo os resultados
print("Grupo 1:")
print(f"  Média: {media_grupo1:.2f}")
print(f"  Mediana: {mediana_grupo1:.2f}")
print(f"  Desvio Padrão: {desvio_padrao_grupo1:.2f}")
print(f"  Variância: {variancia_grupo1:.2f}")

print("\nGrupo 2:")
print(f"  Média: {media_grupo2:.2f}")
print(f"  Mediana: {mediana_grupo2:.2f}")
print(f"  Desvio Padrão: {desvio_padrao_grupo2:.2f}")
print(f"  Variância: {variancia_grupo2:.2f}")

print("\nTeste t de Student:")
print(f"  Estatística t: {t_statistic:.2f}")
print(f"  Valor-p: {p_value:.3f}")

# Interpretando os resultados
alpha = 0.05  # Nível de significância
if p_value < alpha:
    print("\nConclusão: Há uma diferença estatisticamente significativa entre as médias dos grupos.")
else:
    print("\nConclusão: Não há evidências suficientes para concluir que há uma diferença significativa entre as médias dos grupos.")
```

### Exercício 9: Tabela de Frequência para Dados Categóricos
Dado um conjunto de dados categóricos (por exemplo, categorias de produtos), crie:
- Tabela de frequências absolutas
- Tabela de frequências relativas
- Tabela de frequências acumuladas

Para um conjunto de dados numéricos (por exemplo, idades), crie:

- Tabela de frequências por intervalo
- Histograma correspondente à tabela


### Exercício 11: Distribuições de Probabilidade (opcional)
Gere amostras a partir de diferentes distribuições (normal, uniforme, exponencial, etc.) e crie tabelas de frequência para cada uma. Compare as distribuições obtidas com as distribuições teóricas.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribuição Normal
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=10, size=1000)  # Média 50, Desvio Padrão 10
df_normal = pd.DataFrame({'Valor': normal_data})

# Criando a tabela de frequências
frequencia_normal = pd.cut(df_normal['Valor'], bins=20).value_counts(sort=False)
print("Tabela de Frequência - Distribuição Normal:")
print(frequencia_normal)

# Visualizando a distribuição
sns.histplot(normal_data, bins=20, kde=True)
plt.title('Distribuição Normal')
plt.show()

# 2. Distribuição Uniforme
uniform_data = np.random.uniform(low=0, high=100, size=1000)
df_uniform = pd.DataFrame({'Valor': uniform_data})

# Criando a tabela de frequências
frequencia_uniform = pd.cut(df_uniform['Valor'], bins=20).value_counts(sort=False)
print("\nTabela de Frequência - Distribuição Uniforme:")
print(frequencia_uniform)

# Visualizando a distribuição
sns.histplot(uniform_data, bins=20, kde=True)
plt.title('Distribuição Uniforme')
plt.show()

# 3. Distribuição Exponencial
exponential_data = np.random.exponential(scale=5, size=1000)  # Escala (lambda) = 5
df_exponential = pd.DataFrame({'Valor': exponential_data})

# Criando a tabela de frequências
frequencia_exponential = pd.cut(df_exponential['Valor'], bins=20).value_counts(sort=False)
print("\nTabela de Frequência - Distribuição Exponencial:")
print(frequencia_exponential)

# Visualizando a distribuição
sns.histplot(exponential_data, bins=20, kde=True)
plt.title('Distribuição Exponencial')
plt.show()
```

### Exercício 13: Gráficos para Dados Categóricos
Para um conjunto de dados categóricos, crie e interprete:
- Gráfico de barras
- Gráfico de setores (pizza)
- Gráfico de Pareto

```python
# Definindo os gêneros de filmes e a quantidade de filmes em cada gênero
generos = ['Ação', 'Comédia', 'Drama', 'Aventura', 'Ficção Científica', 'Romance', 'Terror']
ocorrencias = [50, 30, 25, 40, 15, 20, 10]  # Número de filmes em cada gênero

# Criando o DataFrame
df_generos = pd.DataFrame({'Gênero': generos, 'Ocorrências': ocorrencias})

# Ordenando por frequência para facilitar o gráfico de Pareto
df_generos = df_generos.sort_values(by='Ocorrências', ascending=False)

# Gráfico de Barras
plt.figure(figsize=(10, 6))
plt.bar(df_generos['Gênero'], df_generos['Ocorrências'], color='skyblue')
plt.title('Gráfico de Barras - Número de Filmes por Gênero')
plt.xlabel('Gêneros')
plt.ylabel('Número de Filmes')
plt.show()

# Gráfico de Setores (Pizza)
plt.figure(figsize=(12, 12))
plt.pie(df_generos['Ocorrências'], labels=df_generos['Gênero'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Gráfico de Setores (Pizza) - Proporção de Filmes por Gênero')
plt.show()

# Gráfico de Pareto
# Frequência acumulada
df_generos['Frequência Acumulada'] = df_generos['Ocorrências'].cumsum()
df_generos['Proporção Acumulada'] = df_generos['Frequência Acumulada'] / df_generos['Ocorrências'].sum()

fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras para as ocorrências
ax1.bar(df_generos['Gênero'], df_generos['Ocorrências'], color='skyblue', alpha=0.7)
ax1.set_xlabel('Gêneros')
ax1.set_ylabel('Número de Filmes', color='skyblue')

# Linha para a proporção acumulada
ax2 = ax1.twinx()
ax2.plot(df_generos['Gênero'], df_generos['Proporção Acumulada'], color='red', marker='o', linestyle='dashed')
ax2.set_ylabel('Proporção Acumulada', color='red')

plt.title('Gráfico de Pareto - Número de Filmes por Gênero e Proporção Acumulada')
plt.show()
```


### Exercício 15: Visualização de Relações
Dado um conjunto de dados com múltiplas variáveis, crie:
- Matriz de correlação
- Scatter plot matrix
- Pairplot


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criando o DataFrame df_vendas
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
num_vendas = 200

np.random.seed(42)  # Para reprodutibilidade
vendas = np.random.randint(1000, 5000, num_vendas)
regioes_vendas = np.random.choice(regioes, num_vendas)

df_vendas = pd.DataFrame({'Região': regioes_vendas, 'Vendas': vendas})

# 1. Matriz de Correlação (apenas para 'Vendas')
correlation_matrix = df_vendas[['Vendas']].corr()
plt.figure(figsize=(11, 11))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação - Vendas')
plt.show()

# 2. Matriz de Gráficos de Dispersão (apenas para 'Vendas')
sns.pairplot(df_vendas[['Vendas']], height=12)
plt.suptitle('Gráfico de Dispersão - Vendas', y=1.02)
plt.show()

# 3. Pairplot (com tonalidade para 'Região')
sns.pairplot(df_vendas, hue='Região', vars=['Vendas'], diag_kind="hist", height=12)
plt.suptitle('Pairplot com Tonalidade por Região', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

# Criando o DataFrame df_vendas
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
num_vendas = 200

np.random.seed(42)  # Para reprodutibilidade
vendas = np.random.randint(1000, 5000, num_vendas)
regioes_vendas = np.random.choice(regioes, num_vendas)

df_vendas = pd.DataFrame({'Região': regioes_vendas, 'Vendas': vendas})

# Definindo as cores fixas para cada região
color_mapping = {
    'Norte': 'blue',
    'Nordeste': 'green',
    'Centro-Oeste': 'red',
    'Sudeste': 'yellow',
    'Sul': 'purple'
}


# 1. Heatmap
heatmap_data = pd.pivot_table(df_vendas, values='Vendas', index='Região', aggfunc='mean')
plt.figure(figsize=(12, 12))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".2f", cbar=False)
plt.title('Heatmap - Média de Vendas por Região')


# Adicionando cores fixas ao heatmap
for i, region in enumerate(regioes):
    heatmap_data.index = heatmap_data.index.astype(str)
    if region in heatmap_data.index:
        plt.gca().get_yticklabels()[heatmap_data.index.get_loc(region)].set_color(color_mapping[region])

plt.show()

# 2. Treemap
treemap_data = df_vendas.groupby('Região')['Vendas'].sum().reset_index()
plt.figure(figsize=(12, 12))
squarify.plot(sizes=treemap_data['Vendas'], label=treemap_data['Região'], alpha=.8,
              color=[color_mapping[region] for region in treemap_data['Região']])
plt.title('Treemap - Total de Vendas por Região')
plt.axis('off')
plt.show()

# 3. Gráfico de Bolhas (com cores mapeadas)
bubble_data = df_vendas.groupby('Região')['Vendas'].agg(['mean', 'std']).reset_index()
bubble_data.columns = ['Região', 'Média', 'Desvio Padrão']
plt.figure(figsize=(10, 6))

for region in bubble_data['Região']:
    region_data = bubble_data[bubble_data['Região'] == region]
    plt.scatter(region_data['Região'], region_data['Média'], s=region_data['Desvio Padrão'] * 100,
                alpha=0.7, color=color_mapping[region])

plt.title('Gráfico de Bolhas - Média e Desvio Padrão de Vendas por Região')
plt.xlabel('Região')
plt.ylabel('Média de Vendas')
plt.show()

# 4. Facetplot (com cores mapeadas)
g = sns.FacetGrid(df_vendas, col='Região', hue='Região', palette=color_mapping, height=4, aspect=1)
g.map(sns.histplot, 'Vendas', kde=True)
g.set_titles("{col_name}") # define os títulos dos subplots como o nome da região
g.fig.suptitle('Facetplot - Distribuição de Vendas por Região', y=1.02)
plt.show()
```
